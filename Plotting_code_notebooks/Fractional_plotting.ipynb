```python
def assign_bins(table,Nb=20,fixedsizes=True):
    
    sort_order = np.argsort(table['x']) # Sort the table by binning column.
    fracs = np.zeros((Nb,3)) # Pre-allocate array with Nb bins.
    
    if fixedsizes is True:
        bin_edges = np.linspace(0,1,Nb+1)
        bin_edges[-1] = 2 # Ensure final value goes into final bin.
        bin_values = np.linspace(0,1,len(table))
        
        bins = np.empty(len(table['x']))
        bins[sort_order] = np.digitize(bin_values,bins=bin_edges)  
    
    else:
        bin_edges = np.linspace(np.min(table['x']),np.max(table['x']),Nb+1)
        bin_edges[-1] = bin_edges[-1]+1 # Ensure final valu goes into final bin.
        bins = np.digitize(table['x'],bins=bin_edges)
        
    return bins
```

```python
def assign(table,th=0.5,redistribute=False,rd_th=0,ct_th=0,print_sizes=False):

    m_columns = ['p_1','p_2','p_3','p_4','p_5','p_ct']
    
    m_array = np.array([table[column] for column in m_columns]).T
    
    arm_assignments = np.ones(len(table))*(-999) # Assigned arm numbers 
    # initially is an array of -999s. -999 means 'no assignment'.
    for m in range(6):
        a = (np.argmax(m_array,axis=1) == m) & (m_array[:,m] >= th)
        arm_assignments[a] = m
        
    if redistribute is True: # Redistribute according to thresholds.
        for m in range(5):
            arm_assignments[(np.argmax(m_array[:,:5],axis=1) == m) 
                & (arm_assignments == 5) & (table[:,m]/table[:,5] > rd_th) 
                & (table[:,5] <= ct_th)] = m

    if print_sizes is True:
        print("total sample: " + str(len(arm_assignments)))
        print("total 'assigned' sample: " 
              + str(np.sum(arm_assignments != -999)))
        for m in range(6):
            print("m = " + str(m+1) + ": " 
                  + str(np.sum(arm_assignments == m)))
      
    return arm_assignments
```

```python
def get_fracs(table,bins,m,arm_assignments,c=0.683):
    
    fracs = np.zeros((np.max(bins),5)) # Pre-assign the fractions array.
    
    for n in range(int(np.max(bins))):
        
        bin_select = (bins == n)
        table_bin = table[bin_select]
        arm_bin = arm_assignments[bin_select]
        
        fracs[n,0] = np.mean(table_bin['x']) # Mean of the 'binned by' parameter 
        # for each bin.
        
        fracs[n,2] = len(table_bin) # Total bin size.
        
        bin_m = table_bin[arm_bin == m]
        fracs[n,1] = len(bin_m) # Number of gals assigned with m arms for 
        # each bin.
        
    for r in range(len(fracs)): # Now get errors according to the Cameron
    # et al. paper:
        n = fracs[r,2] # Number of gals w. m arms.
        k = fracs[r,1] # Number of gals in the bin in total.

        fracs[r,3] = dist.beta.ppf((1-c)/2.,k+1,n-k+1)
        fracs[r,4] = dist.beta.ppf(1-(1-c)/2.,k+1,n-k+1)
        
    fracs = Table(fracs,names=('x','N_m','N_total','f_min','f_max'))
        
    return fracs
```

```python
clr = ['orange','r','purple','g','b']

cx = ['PETROMAG_MR','PETROMAG_MZ']

xticks=[0,0.8,0.2]
xlims=[0,0.8]

xlb = 'r-z'
    
fig,axes = plotting.make_stack(title=False,title_position=[0.1,0.9,20]
                    ,x_limits=xlims,x_label=xlb
                    ,y_label='Fraction',x_ticks=xticks,figsize=(5,8))

table,spirals = load(cx=cx)
a = assign(spirals,print_sizes=True,th=0)
a = a[a != -999] # Only keep gals if they have been 
# assigned an arm number?
b = assign_bins(spirals[a != -999],fixedsizes=True)

for m in range(5):
    
    fracs = get_fracs(table=spirals[a != -999],bins=b
                      ,arm_assignments=a,m=m)
    
    axes[m].plot(fracs['x'],fracs['N_m']/fracs['N_total'],linewidth=2
                 ,color=clr[m])
    
    axes[m].fill_between(fracs['x'],fracs['f_min'],fracs['f_max'],color=clr[m],alpha=0.5)
```
