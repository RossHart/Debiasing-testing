```python
from astropy.cosmology import FlatLambdaCDM,z_at_value
import astropy.units as u

cosmo = FlatLambdaCDM(H0=70,Om0=0.3)

% matplotlib inline
```

```python
data = fits.getdata("../../fits/Full_sample_spec_w_urls.fits",1)

def z_from_mass(mass_lim,mr_limit=17,data=data): # Has the values [lower mass limit,upper mass limit]

    #data = fits.getdata("../../fits/Volume_limited_sample_Baldry_w_bd.fits",1)
    
    colour = data.field("PETROMAG_MU")-data.field("PETROMAG_MR")
    mass = data.field('LOGMSTAR_BALDRY06')
    
    colour_ml = colour[mass > mass_lim]
    colour_ml = np.sort(colour_ml)
    
    colour_percentile = colour_ml[int(0.001*len(colour_ml))-1]
     
    if colour_percentile > 0.79/0.38:
        ML_limit = -0.16 + 0.18*colour_percentile
        
    else:
        ML_limit = -0.95 + 0.56*colour_percentile

    L_limit = (10**mass_lim)/(10**ML_limit)
    Mag_limit = -2.5*math.log10(L_limit) + 4.62
    
    D = 10**(((mr_limit-Mag_limit)/5) + 1)
    z = z_at_value(cosmo.luminosity_distance, (D/10**6) * u.Mpc, zmax=10)
    
    return z # => This is the redshift limit to which we have a stellar mass limited smaple.
```

```python
N_steps = 20

lims = np.empty((N_steps,2))

for mass in enumerate(np.linspace(9,11,N_steps)):
    
    z = z_from_mass(mass[1])
    lims[mass[0],:] = [mass[1],z]
```
